```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
#for time series we are going to use a new library called statsmodel
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# count_per_year_df.index = pd.to_datetime(count_per_year_df.index)
# count_per_year_df = count_per_year_df.asfreq('M')  # Set the appropriate frequency, 'M' for monthly data

count_per_year_df
```

```python


# Assuming count_per_year_df is your time series DataFrame with 'Month' as the index
# Make sure the DataFrame has 'count' as the target variable

# Split the data into training and testing sets
train_size = int(len(count_per_year_df) * 0.8)
train, test = count_per_year_df[:train_size], count_per_year_df[train_size:]

# Fit the ARIMA model
# Here, order=(p, d, q), where p, d, and q are the AR order, differencing order, and MA order, respectively.
order = (3,1, 48)  # Adjust order based on your analysis
model = ARIMA(train['count'], order=order)
results = model.fit()

# Forecast future values
forecast_steps = len(test)
forecast = results.get_forecast(steps=forecast_steps)

# Get confidence intervals
conf_int = forecast.conf_int()

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(train['count'], label='Train')
plt.plot(test['count'], label='Test')
plt.plot(forecast.predicted_mean, label='Forecast', color='red')
plt.fill_between(conf_int.index, conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='red', alpha=0.2)
plt.title('ARIMA Forecast')
plt.legend()
plt.show()
```

```python
forecast.predicted_mean
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Assuming 'test' is your actual test set values and 'forecast' are your predicted values
forecast_values = forecast.predicted_mean

# Convert each individual value to int
forecast_values_int = forecast_values.astype(int)

# Evaluate the model
mse = mean_squared_error(test['count'], forecast_values_int)
mae = mean_absolute_error(test['count'], forecast_values_int)
r2 = r2_score(test['count'], forecast.predicted_mean)

# Print the evaluation metrics
print("R2 value is =", round(r2, 2))
print("The MSE of the model is =", mse)
print("The RMSE of the model is =", np.sqrt(mse))
print("The MAE of the model is =", mae)

```
